```python
import pandas as pd


file_path = 'training_mathbert.xlsx'  
data = pd.read_excel(file_path)

print(data.head())

X_train = data[['embed_2']]  
y_train = data['embed_8']      
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, y_train)

# Predict the embed_8 values for the training set
y_train_pred = reg.predict(X_train)

# Display the coefficients and intercept
print(f'Coefficient: {reg.coef_[0]}')
print(f'Intercept: {reg.intercept_}')
from sklearn.metrics import mean_absolute_error, r2_score


mae = mean_absolute_error(y_train, y_train_pred)
r2 = r2_score(y_train, y_train_pred)

print(f'Mean Absolute Error: {mae}')
print(f'R-squared: {r2}')
import matplotlib.pyplot as plt

plt.scatter(X_train, y_train, color='blue', label='Actual')
plt.plot(X_train, y_train_pred, color='red', label='Predicted', linewidth=2)
plt.xlabel('embed_2') 
plt.ylabel('embed_8')    
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
